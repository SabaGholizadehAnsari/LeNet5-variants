```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

from torchvision import datasets
from torchvision.transforms import ToTensor

from torch import nn


import matplotlib.pyplot as plt
import pandas as pd
from torchvision.io import read_image
import os
import numpy as np

models_path = 'Model'
plots_path = 'Graph'
```

```python
train_data=datasets.FashionMNIST(root="data", train=True,download=True,transform=ToTensor())
test_data=datasets.FashionMNIST( root="data",train=False,download=True,transform=ToTensor())
train_data,valid_data=torch.utils.data.random_split(train_data, [50000, 10000])

train_dataloader = DataLoader(train_data, batch_size=32, shuffle=True)
valid_dataloader = DataLoader(valid_data, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=32, shuffle=True)
len(train_dataloader), len(valid_dataloader), len(test_dataloader)

#train_feature,train_label=next(iter(train_dataloader))
```

#Define *Original* LeNet5

```python
from collections import OrderedDict

class LeNet5(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature = nn.Sequential(OrderedDict([
                  ('conv1', nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=2)),   # 28*28->32*32-->28*28
            ('relu1', nn.ReLU()),
            ('pool1', nn.AvgPool2d(kernel_size=2, stride=2)),  # 14*14

            ('conv2', nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1)),   # 28*28->32*32-->28*28
            ('relu2', nn.ReLU()),
            ('pool2',nn.AvgPool2d(kernel_size=2, stride=2))])  # 14*14
        )
        self.classifier = nn.Sequential(
           OrderedDict([
                  ('flat', nn.Flatten()),
             ('fc1',  nn.Linear(5*5*16, 120)),
            ('relu3', nn.ReLU()),
            ('fc2',  nn.Linear(in_features=120, out_features=84)),
          ('relu4', nn.ReLU()),
            ('fc3',  nn.Linear(in_features=84, out_features=10))])
            )

    def forward(self, x):
        logits = self.classifier(self.feature(x))
        return logits
```

#Network Summary

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")

LeNet5=LeNet5().to(device)
print(LeNet5)

from torchinfo import summary

summary(model=LeNet5, input_size=(1, 1, 28, 28), col_width=20,
                  col_names=['input_size', 'output_size', 'num_params', 'trainable'], row_settings=['var_names'], verbose=0)
```

#Train the LeNet5 (original)

```python
from torchmetrics import Accuracy

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(LeNet5.parameters(), lr = 0.01)
accuracy = Accuracy(task='multiclass', num_classes=10)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
accuracy = accuracy.to(device)
#model_lenet5v1 = LeNet5.to(device)

# Training with Validation
epochs = 15
train_losses = []
valid_losses = []
train_accs=[]
valid_accs=[]
for e in range(epochs):
    train_loss = 0.0
    train_acc=0.0
    for data, labels in train_dataloader:
        # Transfer Data to GPU if available
        if torch.cuda.is_available():
            data, labels = data.cuda(), labels.cuda()


        # Forward Pass
        LeNet5.train()
        target = LeNet5(data)
        # Find the Loss
        loss = criterion(target,labels)
        acc=accuracy(target,labels)
        # Calculate Loss
        train_loss += loss.item()
        train_acc += acc

       # Clear the gradients
        optimizer.zero_grad()
        # Calculate gradients
        loss.backward()
        # Update Weights
        optimizer.step()

    train_loss /= len(train_dataloader)
    train_acc /= len(train_dataloader)

    valid_loss, valid_acc = 0.0, 0.0
    LeNet5.eval()     # Optional when not using Model Specific layer
    for data, labels in valid_dataloader:
        # Transfer Data to GPU if available
        if torch.cuda.is_available():
            data, labels = data.cuda(), labels.cuda()

        # Forward Pass
        target = LeNet5(data)
        # Find the Loss
        loss = criterion(target,labels)
        # Calculate Loss
        valid_loss += loss.item()
        acc=accuracy(target,labels)
        valid_acc += acc

    valid_loss /= len(valid_dataloader)
    valid_acc /= len(valid_dataloader)
    train_losses.append(train_loss)
    valid_losses.append(valid_loss)
    train_accs.append(train_acc)
    valid_accs.append(valid_acc)
    print(f"Epoch: {e}| Train loss: {train_loss: .5f}| Train acc: {train_acc: .5f}| Val loss: {valid_loss: .5f}| Val acc: {valid_acc: .5f}")
```

#Plot the accuracy and loss

```python
def plot_train_val_metrics(train_losses,valid_losses,train_acc,valid_acc, model_name):
    plt.figure(figsize=(10, 4))

    # Plot training and validation loss
    plt.subplot(1, 2, 1)
    plt.plot(train_losses,color='black')
    plt.plot(valid_losses,color='red')
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')

    # Plot training and validation accuracy
    plt.subplot(1, 2, 2)
    plt.plot(train_acc, color='black')
    plt.plot(valid_acc, color='red')
    plt.title('Training & Validation Accuracy')
    plt.legend(['train', 'val'], loc='upper left')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(['train', 'val'], loc='upper left')

    # Add a main title for the whole figure
    plt.suptitle(f"Training and Validation Metrics for {model_name}", fontsize=20, y=1.05)

    # Adjust layout and show plot
    plt.tight_layout()

    # Save plot to file
    save_path = os.path.join(plots_path +'/', f"{model_name}_metrics.png")
    plt.savefig(save_path, bbox_inches='tight', dpi=300)

    plt.show()
```
